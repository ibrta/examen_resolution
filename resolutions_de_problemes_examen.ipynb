```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt #L'importation de matplotlib.pyplot a été ajoutée pour afficher les images.


# Charger les données CIFAR-10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Vérifier les formes des tableaux importés
print('x_train shape:', x_train.shape)
print('y_train shape:', y_train.shape)
print('x_test shape:', x_test.shape)
print('y_test shape:', y_test.shape)

# Afficher un exemple d'image
plt.imshow(x_train[0])
plt.title('Example Image')
plt.axis('off')
plt.show()

# Afficher les étiquettes
labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
print('Label:', labels[y_train[0][0]])

# Normalisation des données
x_train_norm = x_train.astype('float32') / 255.0 #La conversion en float32 a été ajoutée avant la normalisation des données pour éviter des erreurs de type.
x_test_norm = x_test.astype('float32') / 255.0 #La conversion en float32 a été ajoutée avant la normalisation des données pour éviter des erreurs de type.

# Afficher les valeurs normalisées d'un exemple
print('Normalized Pixel Values for Example Image:')
print(x_train_norm[0][0][:3])  # Afficher les trois premières valeurs normalisées du premier pixel

# One-Hot Encoding des étiquettes
y_train_encoded = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test_encoded = tf.keras.utils.to_categorical(y_test, num_classes=10)

# Vérifier les formes des tableaux encodés
print('y_train_encoded shape:', y_train_encoded.shape)
print('y_test_encoded shape:', y_test_encoded.shape)

# Création d'un modèle
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)), #Le padding 'same' a été ajouté aux couches convolutionnelles pour conserver la taille des images
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu', padding='same'), #Le padding 'same' a été ajouté aux couches convolutionnelles pour conserver la taille des images
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu', padding='same'), #Le padding 'same' a été ajouté aux couches convolutionnelles pour conserver la taille des images
    Flatten(),
    Dropout(0.2),
    Dropout(0.2), #Une couche Dropout a été ajoutée après la couche Flatten et avant la dernière couche Dense pour réduire le sur-apprentissage.
    Dense(10, activation='softmax')
])

# Afficher le résumé du modèle
model.summary()

# Compilation du modèle
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entraînement du modèle
history = model.fit(x_train_norm, y_train_encoded, epochs=20, validation_data=(x_test_norm, y_test_encoded)) #Le nombre d'époques a été augmenté à 20 pour un meilleur entraînement.

# Évaluation du modèle
loss, accuracy = model.evaluate(x_test_norm, y_test_encoded)
print('Test set accuracy:', accuracy * 100)
```

```python
import numpy as np #ajout de la librairie numpy

# Afficher les courbes de précision et de perte
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Tester une image et afficher l'image prédite et l'image réelle
def test_image(index):
    test_image = x_test_norm[index]
    test_label = y_test[index]

# Prédiction de l'image
    prediction = model.predict(np.expand_dims(test_image, axis=0))
    predicted_label = np.argmax(prediction)

# Affichage de l'image prédite et de l'image réelle
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(x_test[index])
    plt.title('True Label: {}\nPredicted Label: {}'.format(labels[test_label[0]], labels[predicted_label]))
    plt.axis('off')

# Affichage des probabilités prédites pour chaque classe
    plt.subplot(1, 2, 2)
    plt.bar(range(10), prediction[0])
    plt.title('Predicted Probabilities')
    plt.xlabel('Class')
    plt.ylabel('Probability')
    plt.xticks(range(10), labels, rotation=45)
    plt.show()

# Tester une image (remplacez index par l'indice de l'image que vous voulez tester)
test_image(10)
```
