````python
# Questions de développement - LDD, LMD, LID, et LCD

# Instructions :
Pour chaque question ci-dessous, expliquez le terme demandé et donnez un exemple de commande SQL correspondant.

---
**Question 1 :** Expliquez ce qu'est le LDD (Langage de Définition de Données), son utilité dans la base de données, et donnez quelques exemples de commande LDD (au moins 3).
LDD (Langage de Définition de Données)


Le LDD est un sous-ensemble du langage SQL utilisé pour définir et gérer la structure de la base de données.
Il permet de créer, modifier et supprimer des objets dans la base de données tels que les tables, les vues, les index et les contraintes.
Le LDD est essentiel pour établir et maintenir la structure de la base de données, définir les relations entre les tables, et assurer l'integrité des données.


Exemples de commandes LDD :
CREATE TABLE : pour créer une nouvelle table
ALTER TABLE : pour modifier la structure d'une table existante
DROP TABLE : pour supprimer une table
CREATE INDEX : pour créer un index sur une ou plusieurs colonnes
CREATE VIEW : pour créer une vue
TRUNCATE TABLE : pour vider une table de ses données
---

**Question 2 :** Expliquez ce qu'est le LMD (Langage de Manipulation de Données), son utilité dans la base de données, et donnez quelques exemples de commande LMD (au moins 3).


Le LMD est un sous-ensemble du SQL utilisé pour manipuler les données stockées dans la base de données.
Il permet d'insérer, modifier, supprimer et récupérer des données.
Le LMD est crucial pour interagir avec les données stockées dans la base, permettant aux utilisateurs et aux applications de gérer le contenu de la base de données.

INSERT : pour ajouter de nouvelles données dans une table
UPDATE : pour modifier des données existantes dans une table
DELETE : pour supprimer des données d'une table
SELECT : pour récupérer des données d'une ou plusieurs tables

---

**Question 3 :** Expliquez ce qu'est le LCD (Langage de Contrôle de Données), son utilité dans la base de données, et donnez quelques exemples de commande LCD (au moins 3).


Le LCD est un sous-ensemble du SQL utilisé pour gérer les droits d'accès et les permissions dans la base de données.
Il contrôle qui peut accéder à quelles données et quelles opérations ils peuvent effectuer.
Le LCD est essentiel pour la sécurité de la base de données, permettant de définir et gérer les droits des utilisateurs sur les objets de la base.

GRANT : pour donner des permissions spécifiques à un utilisateur
REVOKE : pour retirer des permissions à un utilisateur
CREATE USER : pour créer un nouvel utilisateur dans la base de données
ALTER USER : pour modifier les paramètres d'un utilisateur existant
DROP USER : pour supprimer un utilisateur de la base de données


---

**Question 4 :** Expliquez ce qu'est le LID (Langage d'interrogation de Données), son utilité dans la base de données, et donnez quelques exemples de commande LID (au moins 3).

Le LID, est généralement considéré comme faisant partie du LMD.
Il se concentre spécifiquement sur la récupération des données.
Le LID permet aux utilisateurs et aux applications d'extraire des informations spécifiques de la base de données selon des critères définis.

SELECT ... FROM : pour spécifier les colonnes à récupérer d'une ou plusieurs tables
SELECT ... WHERE : pour filtrer les résultats selon des conditions spécifiques
SELECT ... ORDER BY : pour trier les résultats
SELECT ... GROUP BY : pour regrouper les résultats
SELECT ... HAVING : pour filtrer les résultats après un regroupement
````
